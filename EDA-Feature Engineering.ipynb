```python
# Example if your file is in a folder called "data" on your desktop:
file_path = r'D:\DOWNLOADS\Country-Code.xlsx'

# The 'r' before the string is important! It makes it a "raw string" 
# so you don't have to worry about escaping the backslashes (\).

# Then use the path in your pandas call:
df_country= pd.read_excel(file_path)
df_country.head()
```

```python
country_names=final_df.Country.value_counts().index
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating", y="Rating Count",  data=ratings)
```

```python
sns.barplot(x="Aggregate rating", y="Rating Count", hue='Rating color' ,data=ratings)
```

```python
#mapping the color
sns.barplot(x="Aggregate rating", y="Rating Count", hue='Rating color' ,data=ratings, palette=['white', 'red','orange','yellow','green','green'])
```
