```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.33, random_state=42)
```

```python
##linear regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lin_reg=LinearRegression()
##5-cross validation
mse=cross_val_score(lin_reg,X_train,y_train,scoring='neg_mean_squared_error',cv=5)
mean_mse=np.mean(mse)
print(mse)
print(mean_mse)
```

```python
#Ridge regresssion
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge=Ridge()
params={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20]} #try all params and  gives best fit params
ridge_regressor=GridSearchCV(ridge,params,scoring='neg_mean_squared_error',cv=5)
ridge_regressor.fit(X_train,y_train)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

```python
##we cannot use ridge because performace is getting more bad
##so we will use lasso
```

```python
#Lassso regresssion
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

lasso=Lasso()
params = {'alpha': [1e-15, 1e-10, 1e-8, 1e-3, 1e-2, 1, 5, 10, 20]} #try all params and  gives best fit params
lasso_regressor = GridSearchCV(lasso, params, scoring='neg_mean_squared_error', cv=5,verbose=3)
lasso_regressor.fit(X_train, y_train)
```

```python
# The optimal alpha value found by the grid search
print(f"Best Alpha: {lasso_regressor.best_params_}")
```

```python
# 1. Best Alpha Value
print("--- Best Hyperparameter ---")
print(f"Best Alpha: {lasso_regressor.best_params_}")

# 2. Best Model Score
print("\n--- Best Score ---")
print(f"Best Score (Negative MSE): {lasso_regressor.best_score_}")
print(f"Best Mean Squared Error (MSE): {-lasso_regressor.best_score_}")

# 3. Accessing the full results (all 45 runs)
print("\n--- Full Results Overview ---")
results = pd.DataFrame(lasso_regressor.cv_results_)
# This DataFrame shows the score for every alpha in every fold
print(results[['param_alpha', 'mean_test_score', 'rank_test_score']].head(9))
```
